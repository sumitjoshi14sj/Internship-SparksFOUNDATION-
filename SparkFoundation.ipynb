```python
#Sales CountPLot
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
sns.set_style('whitegrid')
#sns.countplot(x='State',data=df)
#to make them in order
sns.countplot(df['State'],order=df['State'].value_counts().index)
```

```python
# Sub-Category Wise countplot
plt.figure(figsize=(20,9))
plt.xticks(rotation=90)
#sns.countplot(x='Sub-Category',data=df)
#In order
sns.countplot(df['Sub-Category'],order=df['Sub-Category'].value_counts().index)
```

```python
#Segmnent wise countplot
sns.countplot(x='Segment',data=df)
```

```python
#PLot of subcategory and sales
sns.set_style('whitegrid')
plt.xticks(rotation=90)
sns.barplot(x=df['Sub-Category'],y=df['Sales'])
```

```python
#PLot of sates and  profit
plt.figure(figsize=(20,9))
sns.set_style('whitegrid')
plt.xticks(rotation=90)
sns.barplot(x=df['State'],y=df['Profit'])
```

```python
# State wise sales
plt.figure(figsize=(20,9))
sns.set_style('whitegrid')
plt.xticks(rotation=90)
sns.barplot(x=df['State'],y=df['Sales'])
```

```python
#Top 10 state wise Sales

top10cities= df.groupby('State')['Sales'].sum().sort_values(ascending=False).iloc[0:10]
plt.title('Sales of Top 10 States')
plt.xlabel('State')
plt.ylabel('Total sum of sales')
top10cities.plot(kind='bar')
```

```python
#Top 10 cities wise sales
top10cities= df.groupby('City')['Sales'].sum().sort_values(ascending=False).iloc[0:10]
plt.title('Sales of Top 10 Cities')
plt.xlabel('City')
plt.ylabel('Total sum of sales')
top10cities.plot(kind='bar')
```

```python
discount= df.groupby('City')['Discount'].value_counts().sort_values(ascending=False).iloc[0:10]
discount
```
